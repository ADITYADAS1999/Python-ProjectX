```python
import random
import time

def triangular_encrypt_block(bit_stream, option):
    triangle = [list(bit_stream)]
    triangle_output = [f"Iteration-1 {' '.join(list(bit_stream))}"]

    for iteration_index in range(len(bit_stream) - 1):
        new_iteration = []
        for j in range(len(triangle[iteration_index]) - 1):
            new_bit = str(int(triangle[iteration_index][j]) ^ int(triangle[iteration_index][j + 1]))
            new_iteration.append(new_bit)
        triangle.append(new_iteration)
        triangle_output.append(f"Iteration-{iteration_index + 2} {' '.join(new_iteration)}")

    if option == '001':
        encrypted_stream = ''.join(row[0] for row in triangle)
    elif option == '010':
        encrypted_stream = ''.join(row[0] for row in reversed(triangle))
    elif option == '011':
        encrypted_stream = ''.join(row[-1] for row in triangle)
    else:  # '100'
        encrypted_stream = ''.join(row[-1] for row in reversed(triangle))

    return encrypted_stream, triangle_output

def get_blocks_from_file(file_path, is_target_file=True):
    blocks = []
    with open(file_path, 'r') as f:
        for line in f:
            try:
                if is_target_file:
                    block = line.strip().split('|')[0].split(': ')[1].strip()
                else:
                    block = line.strip()
                blocks.append(block)
            except IndexError:
                print(f"Skipping line due to unexpected format: {line.strip()}")
    return blocks

def process_blocks_and_compare(target_block_file, input_block_file):
    target_blocks = get_blocks_from_file(target_block_file, is_target_file=True)
    input_blocks = get_blocks_from_file(input_block_file, is_target_file=False)

    with open('/content/drive/MyDrive/SourceTargetBlock.txt', 'w') as f:
        for i, binary_block in enumerate(target_blocks):
            print(f"Processing target block: {binary_block}")
            for option in ['001', '010', '011', '100']:
                encrypted_stream, _ = triangular_encrypt_block(binary_block, option)
                print(f"Option {option}: Resulting encrypted stream: {encrypted_stream}")

                if encrypted_stream in input_blocks:
                    text_character = chr(int(encrypted_stream, 2))  # Convert binary to ASCII character
                    f.write(f"{encrypted_stream} : Text Character: '{text_character}'\n")
                    print(f"Match found with option {option}: Target Block: {binary_block} | Input Block: {encrypted_stream}\n")
                    break
                else:
                    print("No match found for this option.\n")

# File paths - Update these paths to match your file locations
target_block_file_path = '/content/drive/MyDrive/TargetBlock.txt'
input_block_file_path = '/content/drive/MyDrive/InputBlock.txt'

try:
    process_blocks_and_compare(target_block_file_path, input_block_file_path)
except Exception as e:
    print(f"An error occurred: {e}")






# Record start time
start_time = time.time()

# Execute the main process
try:
    process_blocks_and_compare(target_block_file_path, input_block_file_path)
except Exception as e:
    print(f"An error occurred: {e}")

# Record end time
end_time = time.time()

# Calculate total time taken in seconds
total_time = end_time - start_time

print(f"Overall Decryption time taken: {total_time:.2f} seconds")
```

```python
import random
import time

def triangular_encrypt_block(bit_stream, option):
    triangle = [list(bit_stream)]
    triangle_output = [f"Iteration-1 {' '.join(list(bit_stream))}"]

    for iteration_index in range(len(bit_stream) - 1):
        new_iteration = []
        for j in range(len(triangle[iteration_index]) - 1):
            new_bit = str(int(triangle[iteration_index][j]) ^ int(triangle[iteration_index][j + 1]))
            new_iteration.append(new_bit)
        triangle.append(new_iteration)
        triangle_output.append(f"Iteration-{iteration_index + 2} {' '.join(new_iteration)}")

    if option == '001':
        encrypted_stream = ''.join(row[0] for row in triangle)
    elif option == '010':
        encrypted_stream = ''.join(row[0] for row in reversed(triangle))
    elif option == '011':
        encrypted_stream = ''.join(row[-1] for row in triangle)
    else:  # '100'
        encrypted_stream = ''.join(row[-1] for row in reversed(triangle))

    return encrypted_stream, triangle_output

def get_blocks_from_file(file_path, is_target_file=True):
    blocks = []
    with open(file_path, 'r') as f:
        for line in f:
            try:
                if is_target_file:
                    block = line.strip().split('|')[0].split(': ')[1].strip()
                else:
                    block = line.strip()
                blocks.append(block)
            except IndexError:
                print(f"Skipping line due to unexpected format: {line.strip()}")
    return blocks

def process_blocks_and_compare(target_block_file, input_block_file):
    target_blocks = get_blocks_from_file(target_block_file, is_target_file=True)
    input_blocks = get_blocks_from_file(input_block_file, is_target_file=False)

    with open('/content/drive/MyDrive/SourceTargetBlock.txt', 'w') as f:
        for i, binary_block in enumerate(target_blocks):
            print(f"Processing target block: {binary_block}")
            for option in ['001', '010', '011', '100']:
                encrypted_stream, _ = triangular_encrypt_block(binary_block, option)
                print(f"Option {option}: Resulting encrypted stream: {encrypted_stream}")

                if encrypted_stream in input_blocks:
                    text_character = chr(int(encrypted_stream, 2))  # Convert binary to ASCII character
                    f.write(f"{encrypted_stream} : Text Character: '{text_character}'\n")
                    print(f"Match found with option {option}: Target Block: {binary_block} | Input Block: {encrypted_stream}\n")
                    break
                else:
                    print("No match found for this option.\n")

# File paths - Update these paths to match your file locations
target_block_file_path = '/content/drive/MyDrive/TargetBlock.txt'
input_block_file_path = '/content/drive/MyDrive/InputBlock.txt'

try:
    process_blocks_and_compare(target_block_file_path, input_block_file_path)
except Exception as e:
    print(f"An error occurred: {e}")






# Record start time
start_time = time.time()

# Execute the main process
try:
    process_blocks_and_compare(target_block_file_path, input_block_file_path)
except Exception as e:
    print(f"An error occurred: {e}")

# Record end time
end_time = time.time()

# Calculate total time taken in seconds
total_time = end_time - start_time

print(f"Overall Decryption time taken: {total_time:.2f} seconds")



def extract_text_character(input_file_path, output_file_path):
    extracted_characters = ""  # Initialize an empty string to accumulate the extracted characters
    with open(input_file_path, 'r') as input_file:
        for line in input_file:
            # Find the index of the first single quote
            start_quote_index = line.find("'")
            if start_quote_index != -1:
                # Find the index of the second single quote after the first one
                end_quote_index = line.find("'", start_quote_index + 1)
                if end_quote_index != -1:
                    # Extract the text character between the single quotes
                    text_character = line[start_quote_index + 1:end_quote_index]
                    extracted_characters += text_character  # Accumulate the extracted character
    # Write the accumulated string to the output file
    with open(output_file_path, 'w') as output_file:
        output_file.write(extracted_characters)

# File paths
source_target_block_file_path = '/content/drive/MyDrive/SourceTargetBlock.txt'
final_output_file_path = '/content/drive/MyDrive/FinalOutput.txt'

# Call the function to extract text characters and store the whole sentence in FinalOutput.txt
extract_text_character(source_target_block_file_path, final_output_file_path)

print("Sentence extracted and stored in FinalOutput.txt successfully.")


```
